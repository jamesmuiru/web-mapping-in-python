```python
import folium
mount_kenya_counties_with_latlong = [
    {"county": "Meru", "latitude": 0.0470, "longitude": 37.6498},
    {"county": "Tharaka-Nithi", "latitude": -0.2971, "longitude": 38.0166},
    {"county": "Embu", "latitude": -0.5373, "longitude": 37.4578},
    {"county": "Kirinyaga", "latitude": -0.5383, "longitude": 37.3481},
    {"county": "Nyeri", "latitude": -0.4296, "longitude": 36.9434},
    {"county": "Laikipia", "latitude": 0.3774, "longitude": 37.0135}
]

# A map centered around mount kenya
mount_kenya_map = folium.Map(location=[0.152, 37.308], zoom_start=8)

# Add using for loop to add markers on my map
for county in mount_kenya_counties_with_latlong:
    folium.Marker(
        location=[county["latitude"], county["longitude"]],
        popup=county["county"]
    ).add_to(mount_kenya_map)
folium.TileLayer('Stamen Terrain').add_to(mount_kenya_map)
folium.TileLayer('Stamen Toner').add_to(mount_kenya_map)
folium.TileLayer(tiles='https://{s}.tile.openstreetmap.org/{z}/{x}/{y}.png',
                 attr='OpenStreetMap').add_to(mount_kenya_map)
folium.TileLayer(tiles='http://{s}.tile.opentopomap.org/{z}/{x}/{y}.png',
                 attr='OpenTopoMap').add_to(mount_kenya_map)
folium.TileLayer(tiles='https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{x}/{y}',
                 attr='Esri', name='Esri Satellite').add_to(mount_kenya_map)


# saving my map
mount_kenya_map.save("mount_kenya_counties_map.html")
folium.LayerControl().add_to(mount_kenya_map)

# Display of my map
mount_kenya_map
```
